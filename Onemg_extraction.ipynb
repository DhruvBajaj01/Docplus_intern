```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup as soup
from urllib.request import Request,urlopen
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
```

```python
#to get data from csv
df1=pd.read_csv('20-40krecords.csv')
```

```python
#to get URL into list
links=df1['URL']
```

```python
#from webdriver_manager.chrome import ChromeDriverManager

#driver = webdriver.Chrome('D:\Software\chromedriver_win32\chromedriver.exe')
driver = webdriver.Chrome(ChromeDriverManager().install())
```

Variable Initialisation

```python
medname=[]
price=[]
salt=[]
manufacturer=[]
address=[]
required=[]
side_effects=[]
side_effects_1=[]
stock=[]
uses=[]
directions=[]
packsize=[]
unitofmeasurement=[]
units=[]
packform=[]
precaution=[]
url=[]
how_to_use=[]
how_to_cope_se=[]
how_works=[]
benefits=[]
forgot_to_take=[]
quick_tips=[]
introduction=[]
```

Main Extraction

```python
maincount_2=1
for i in range(500):
    side_effects_1=[]
    r=[]
    uses_3=[]
    qutip1=[]
    count_12=0
    count_14=0
    count_15=0
    count_16=0
    #converts the page from dynamic HTML
    driver.get(links[i])
    res=driver.execute_script("return document.documentElement.outerHTML")
    #HTML Parser
    soup_1=soup(res,'lxml')
    
    #Medicine Name,Precaution,how to use, how it works
    if soup_1.find('div',{'class':'DrugHeader__left___19WY-'})==None:
        medname.append('Not Available')
        precaution.append('Not Available')
    else:
        medname.append(soup_1.find('div',{'class':'DrugHeader__left___19WY-'}).get_text())
        big=soup_1.findAll("div",{"class":"DrugOverview__container___CqA8x"})
        big1=soup_1.findAll("h2")
        count_1=0
        j=0
        for he in big1:
            if he.text=="Safety Advice":
                count_1=j
                big4=big[count_1-1].findAll("div",{"class":"DrugOverview__content___22ZBX"})
                count=len(big4)
                r=big4[count-6].text+"The previous suggestion is for Alcohol."+big4[count-5].text+"The previous suggestion is for Pregnant Ladies."+big4[count-4].text+"The previous suggestion is for Breastfeeding."+big4[count-3].text+"The previous suggestion is for Driving."+big4[count-2].text+"The previous suggestion is for Kidney."+big4[count-1].text+"The previous suggestion is for Liver."
                precaution.append(r)
            
            if "use" in he.text:
                count_2=j
                count_12=1
                how_to_use.append(big[count_2-1].find("div",{"class":"DrugOverview__content___22ZBX"}).text)
            if "works" in he.text:
                count_4=j
                count_14=1
                how_works.append(big[count_4-1].find("div",{"class":"DrugOverview__content___22ZBX"}).text)
            if "Benefits" in he.text:
                count_16=1
            if "forget" in he.text:
                count_5=j
                count_15=1
                forgot_to_take.append(big[count_5-1].find("div",{"class":"DrugOverview__content___22ZBX"}).text)
            j=j+1
            
    if count_12!=1:
        how_to_use.append('Not Available')
    if count_14!=1:
        how_works.append('Not Available')
    if count_15!=1:
        forgot_to_take.append('Not Available')
    if count_16!=1:
        benefits.append('Not Available')
    else:
        ben_1=soup_1.find("div",{"class":"ShowMoreArray__tile___2mFZk"})
        ben=ben_1.findAll('div')
        benefits.append(ben[len(ben)-1].text)
    
    #Introduction
    if soup_1.find("div",{"class":"DrugOverview__content___22ZBX"})==None:
        introduction.append('Not Available')
    else:
        introduction.append(soup_1.find("div",{"class":"DrugOverview__content___22ZBX"}).get_text())
    
    #Stock,Price,Form of Medicine etc
    if soup_1.find('div',{'class':'DrugPriceBox__best-price___32JXw'})==None:
        price.append('Not Available')
        packsize.append('Not Available')
        unitofmeasurement.append('Not Available')
        units.append('Not Available')
        packform.append('Not Available')
        stock.append('Not Available')
    else:
        price.append(soup_1.find('div',{'class':'DrugPriceBox__best-price___32JXw'}).get_text())
        quantity=soup_1.find('div',{'class':'DrugPriceBox__quantity___2LGBX'}).get_text()
        a=quantity.split(" ")
        packsize.append(a[0])
        unitofmeasurement.append(a[len(a)-4])
        units.append(a[len(a)-2])
        packform.append(a[len(a)-1])

        if soup_1.find('div',{'class':'style__interaction___3cb12'})==None:
            stock.append('No')
        else:
            stock.append('Yes')
    
    #address of manufacturer
    if soup_1.find("div",{"class":"DrugPage__manufacturer-address___2ACya"})==None:
        address.append("Not Available")
    else:
        address.append(soup_1.find("div",{"class":"DrugPage__manufacturer-address___2ACya"}).get_text())
                   
    #salt used in medicine
    if soup_1.find('div',{'class':'saltInfo DrugHeader__meta-value___vqYM0'})==None:
        salt.append('Not Available')
    else:            
        salt.append(soup_1.find('div',{'class':'saltInfo DrugHeader__meta-value___vqYM0'}).get_text())
    
    #manufacturer
    if soup_1.find('div',{'class':'DrugHeader__meta-value___vqYM0'})==None:                      
        manufacturer.append('Not Available')
    else:               
        manufacturer.append(soup_1.find('div',{'class':'DrugHeader__meta-value___vqYM0'}).get_text())
    
    #prescription
    if soup_1.find("div",{"class":"DrugHeader__prescription-req___34WVy"})==None:
        required.append("Not Available")
    else:
        required.append("Yes")
    
    #sideeffects    
    if soup_1.find('div',{'class':'DrugOverview__list-container___2eAr6 DrugOverview__content___22ZBX'})==None:
        side_effects.append('Not Available')
    else:
        x1=soup_1.find('div',{'class':'DrugOverview__list-container___2eAr6 DrugOverview__content___22ZBX'})
        x2=x1.findAll("li")
        for z in range(len(x2)):
            side_effects_1.append(x2[z].text)
        side_effects.append(','.join(side_effects_1))
    
    #how to cope side effects
    if soup_1.find("ul",{"class":"DrugOverview__left-margin___31s5E"})==None:
        how_to_cope_se.append('Not Available')
    else:
        how_to_cope_se.append(soup_1.find("ul",{"class":"DrugOverview__left-margin___31s5E"}).text)

    
    #Quick Tips
    if soup_1.find('div',{'class':'ExpertAdviceItem__content___1Djk2'})==None:
        quick_tips.append('Not Available')
    else:
        qutip=soup_1.find('div',{'class':'ExpertAdviceItem__content___1Djk2'})
        qutip_1=qutip.findAll('li')
        for q in range(len(qutip_1)):
            qutip1.append(qutip_1[q].text)
        quick_tips.append(','.join(qutip1))
    
    #uses
    if soup_1.find("ul",{"class":"DrugOverview__list___1HjxR DrugOverview__uses___1jmC3"})==None:
        uses.append('Not Available')
    else:
        uses_1=soup_1.find("ul",{"class":"DrugOverview__list___1HjxR DrugOverview__uses___1jmC3"})
        uses_2=uses_1.findAll("li")
        for s in range(len(uses_2)):
            uses_3.append(uses_2[s].text)
        uses.append(','.join(uses_3))
    
    #links
    url.append(links[i])
    
    #to keep check on number of records completed
    print(maincount_2)
    maincount_2+=1
    
```

Data Conversion

```python
column=["Product Name","Marketer Name","Marketer Address","Prescription Required","MRP","Ingredients","Units in Pack","Pack Size","Unit of Measurement","Pack Form","Stock","Introduction","Uses","Benefits","How to Use",
"How medicine works",
"If you forget to take",
"Quick Tips","Side Effects","How to cope with side effects","Precaution","URL"]
```

```python
df=pd.DataFrame(columns=column)
```

```python
df["Product Name"]=pd.Series(medname)
df["Marketer Name"]=pd.Series(manufacturer)
df["Marketer Address"]=pd.Series(address)
df["Prescription Required"]=pd.Series(required)

df["MRP"]=pd.Series(price)
df["Ingredients"]=pd.Series(salt)
df["Units in Pack"]=pd.Series(units)
df["Pack Size"]=pd.Series(packsize)
df["Unit of Measurement"]=pd.Series(unitofmeasurement)
df["Pack Form"]=pd.Series(packform)
df["Stock"]=pd.Series(stock)

df["Introduction"]=pd.Series(introduction)
df["Uses"]=pd.Series(uses)
df["Benefits"]=pd.Series(benefits)
df["How to Use"]=pd.Series(how_to_use)
df["How medicine works"]=pd.Series(how_works)
df["If you forget to take"]=pd.Series(forgot_to_take)


df["Side Effects"]=pd.Series(side_effects)
df["How to cope with side effects"]=pd.Series(how_to_cope_se)
df["Quick Tips"]=pd.Series(quick_tips)
df["Precaution"]=pd.Series(precaution)

df["URL"]=pd.Series(url)
```
